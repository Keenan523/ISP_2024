```python
import pandas as pd
pd.set_option('display.max_columns', 20)
train = pd.read_csv("/kaggle/input/train-test/train_set.csv")
test = pd.read_csv("/kaggle/input/train-test/test_set.csv")
```

```python
train
```

```python
X_train = train.drop(columns = ["datetime", "row_id", "prediction_unit_id", "data_block_id", "target", "county","origin_datetime", "hours_ahead", "product_type"])
y_train = train["target"]
X_test = test.drop(columns = ["datetime", "row_id", "prediction_unit_id", "data_block_id", "target", "county","origin_datetime", "hours_ahead", "product_type"])
y_test = test["target"]
```

```python
print(X_train.isna().sum()) #check for no NA
print(X_train.dtypes) #check for correct data type
```

```python
from sklearn.tree import DecisionTreeRegressor
rg = DecisionTreeRegressor() #regressor w default settings
rg.fit(X_train, y_train) #fitting the X and y training sets with the regression
```

```python
y_predict = rg.predict(X_test) #predicting y with X_test set
#very simple method that can lead to overfitting
y_predict
```

```python
#use mean absolute error(MAE) to evaluate prediction
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, y_predict) #find the MAE between the y_test(actual) and y_predict(predicted) target values
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
#similar process but w random forest
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
```

```python
y_predict_rf = rf.predict(X_test)
```

```python
mean_absolute_error(y_test, y_predict_rf)
```

```python
def negative(list):
    list2 = []
    for i in list:
        if i < 0:
            list2.append(i)
    return list2
len(negative(y_train))
```

```python
import math
rf = RandomForestRegressor()
y_trainlog = []
for i in y_train:
    i2 = i +1
    log = math.log(i2)
    y_trainlog.append(log)
rf.fit(X_train, y_trainlog)
```

```python
y_predict_rflog = rf.predict(X_test)
```

```python
import math
y_predict_rf2 = []
for i in y_predict_rflog:
    normal = math.exp(i)
    y_predict_rf2.append(normal)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, y_predict_rf2)
```
